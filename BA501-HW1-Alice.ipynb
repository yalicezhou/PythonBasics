2. How will you determine the overall acceptance rate of requests?

```python
friend_request = pd.DataFrame(
    {"requester_id":['RequesterID1','RequesterID2','RequesterID3','RequesterID4','RequesterID5'],
     "time":["2017-07-01","2017-07-02","2017-07-02","2017-07-03","2017-07-03"],
     "sent_to_id":['AccepterID1','AccepterID2','AccepterID3','AccepterID4','AccepterID5']},
    columns = ["requester_id", "time", "sent_to_id"]
)

request_accepted = pd.DataFrame(
    {"requester_id":['RequesterID1','RequesterID3','RequesterID4'],
     "time":["2017-07-01","2017-07-02","2017-07-03"],
     "acceptor_id":['AccepterID1','AccepterID3','AccepterID4']},
    columns = ["requester_id", "time", "acceptor_id"]
)

Unique_friend_request = friend_request.drop_duplicates()
Unique_request_accepted = request_accepted.drop_duplicates()

#print(friend_request)
#print(request_accepted)

# (1) Python
Accept_rate = pd.merge(Unique_friend_request,Unique_request_accepted,how="left",on="requester_id")
len(Accept_rate["acceptor_id"][Accept_rate.acceptor_id > 0])/(len(Accept_rate["acceptor_id"])*1.0)
# (2) SQL
# SELECT COUNT(DISTINCT Req.requester_id)/COUNT(DISTINCT Acp.requester_id) AS AcceptanceRate
# FROM Unique_friend_request Req
# INNER JOIN Unique_request_accepted Acp
# ON Req.requester_id=Acq.requester_id AND Req.sent_to_id=Acq.acceptor_id
```
